<h2> <a id="A"> A:RESEARCH QUESTION </a> </h2>

To what extent does Departure Delay in Minutes affect airline passengers’ satisfaction?

<h3> <a id=""> Context</a> </h3>

WGU airlines would like to study the Airline Passenger Satisfaction survey data. The dataset is a result of a survey from passengers traveling with different airlines. The Customer Relations department will benefit from the creation of a logistic regression model that will predict if a customer is satisfied based on survey questions. The study will consist of identifying the best predictor variables to the satisfaction target. We will study the relationship between predictors and to the satisfaction target variable. We will look for the coefficient the logistic regression, the std err and the p-value to see how significant the ‘Departure Delay in Minutes’ is in affecting the satisfaction of the passengers.

<h3> <a id=""> Hypothesis </a> </h3>

Delays is flight departure could be a pain for travelers. Travelers get irritiated and become unsatisfied with whole experience regardless if the food was good and aircarft is clean and stuff are friendly. WGU Airlines would like to understand to what extend does the delay in departure affect the customers satisfaction. Therefore, our alternative hypothesis $H_{0}$: The Departure Delay in Minutes statistically significantly affects the passengers’ satisfaction.

<h4> <a id=""> Define find_outliers function to identify the outliers </a> </h4>

```python
def find_outliers(df, col):
    '''Identifies observations with outliers. function takes df, and col as input'''
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_band = Q1 - 1.5 * IQR
    upper_band = Q3 + 1.5 * IQR
    Index_with_outliers = df.index[(df[col] < lower_band) |   (df[col] > upper_band)]
    
    return Index_with_outliers
```

<h4> <a id=""> Identifying observations indexes with outliers. </a> </h4>

```python
variableswithoutliers =['Departure Delay in Minutes','Arrival Delay in Minutes']
index_list=[]
for col in variableswithoutliers:
     index_list.extend(find_outliers(df, col=col))
print ('Number of observations with outliers: ', str(len(index_list)))  
```

<h4> <a id=""> Index values when there over three distinct values </a> </h4>

```python
def create_dict(list_):
    city_Re = dict()
    j=0
    for i in list_: 
        city_Re[i]= j
        j=j+1
    return city_Re
```

```python
list_ = object_distinct_dict['satisfaction']
satisfaction =create_dict(list_)
satisfaction
```

```python
# Create a dictionary that holds distinct values for each variable with an index
dict_index_dict= dict()
for key , values in object_distinct_dict.items():
    dict_index_dict[key]=create_dict(values)
dict_index_dict   
```

```python
# Converting the distinct values to 1 or 0 (Re-Expression)
for col in dict_index_dict.keys():
    df[col].replace(dict_index_dict[col] , inplace=True)
df[dict_index_dict.keys()]
```

<h3> <a id="">  Split data into Predictor and Predcited datasets  </a> </h3>

```python
list_predictors = sorted([col for col in df.columns if col !='satisfaction'])
print(list_predictors)
```

<h4> <a id="">  Predicted (Target) dataset </a> </h4>

Let's split the predicted variable from the predictors.

```python
y = df['satisfaction']
y 
```

<h4> <a id="">  Features (Predictors) dataset </a> </h4>

```python
X = df[list_predictors]
X
```

<h3> <a id="">  Splitting the data into training and testing datasets </a> </h3>

```python
#Spliting the data into training and testing subsets randomly.
X_train, X_test, y_train, y_test = train_test_split(X, y,stratify=y,test_size = 0.3, random_state=30)
```

```python
# Print shape of the 4 data sets, training and testing
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

<h3> <a id=""> Feature Scaling </a> </h3>

Let's scale the features using StandardScaler, the scaler will change the values to vlaues between -1 and 1. This step is necessary for machine learning. Standardizing data gives equal weights to all features during the fitting of the machine learning model.

```python
#initializing the StandardScaler
scaler = StandardScaler()
```

```python
# Standardizing the training dataset
X_train_copy = pd.DataFrame(scaler.fit_transform(X_train))
X_train_copy.columns = X_train.columns.values
X_train_copy.index = X_train.index.values
X_train = X_train_copy
```

```python
# Standardizing the testing dataset
X_test_copy = pd.DataFrame(scaler.transform(X_test))
X_test_copy.columns = X_test.columns.values
X_test_copy.index = X_test.index.values
X_test = X_test_copy
```

```python
#Displaying top 5 records of the standardized testing dataset
X_test.head()
```

```python
#Displaying top 5 records of the standardized training dataset
X_train.head()
```

<h3> <a id=""> Logistic Regression - Initial model </a> </h3>

<h4> <a id=""> Fitting initial Model </a> </h4>

```python
initial_model = sm.Logit(endog=y_train, exog=X_train).fit()
```

```python
# print summary of the initial model
print(initial_model.summary())
```

Notice below that we rounded the pred_y so we can get 0 or 1s. Logistic Regression Model uses a sigmoid function for prediction with results between 0 and 1


```python
# get the predicted values for the test dataset [0, 1]
pred_initial = initial_model.predict(exog=X_test)
pred_y = [int(x) for x in round(pred_initial)]
pred_y[:5]
```
