## B1:CODE

```python
# importing statsmodels methodes to do the ANOVA calculations.
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
#creating variables gender and tenure to compare with ANOVA.
gender_ = df['Gender']
tenure_ = df['Tenure']
```

```python
#Creating Model and generating the p-value and F statistic and degree of freedom (df) and other metrics.
model = ols('tenure_ ~ gender_',
           data = df).fit()

anova_result = sm.stats.anova_lm(model, typ=2)
anova_result
```

We accept the Null hypthesis because there is not enough statistical evidence based ont he p-value.

### C:UNIVARIATE STATISTICS

##### 	Tenure (Continuous) 

```python
sns.set_theme(style="ticks")

f, ax = plt.subplots(figsize=(7, 5))
sns.despine(f)

sns.histplot(
    df,
    x="Tenure",
    multiple="stack",
    palette="light:m_r",
    edgecolor=".3",
    linewidth=.5,
    log_scale=False,
)
plt.title('Tenure  Distribution')
plt.show()
```
